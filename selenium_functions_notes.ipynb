## 2. Locating Elements
- `find_element` finds the first matching element
- `find_elements` finds all matching elements
- `By.ID`, `By.CLASS_NAME`, `By.NAME`, `By.TAG_NAME`, `By.CSS_SELECTOR`, `By.XPATH` specify search methods

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.edge.service import Service
from selenium import webdriver

EDGE_DRIVER_PATH = 'C:/Users/boume/Downloads/Compressed/edgedriver_win32/msedgedriver.exe'

service = Service(executable_path=EDGE_DRIVER_PATH)

driver = webdriver.Edge(service=service)

driver.get('https://fbref.com/en/comps/9/2024-2025/2024-2025-Premier-League-Stats')
driver.implicitly_wait(3)
print(f"Collecting data from fbref: \n")
TABLE_ID = 'results2024-202591_overall'
data = []
squad_links = []
try:
    main_table_links = driver.find_elements(
        By.CSS_SELECTOR, f'#{TABLE_ID} tbody a'
    )
    for link in main_table_links:
        href = link.get_attribute('href')
        if href and 'squads' in href: 
            squad_links.append(href)
except Exception as e :
    print(e)
finally :
    print(squad_links)



try :
    
    table = driver.find_element(By.ID,TABLE_ID)
    print('getting the table ... ')
    rows = table.find_elements(By.TAG_NAME , 'tr')
    for row in rows :
            
        cells = row.find_elements(By.TAG_NAME , 'td')+row.find_elements(By.TAG_NAME, 'td')
    
        row_data = [cell.text.strip() for cell in cells if cell]
        print(row_data)
        if row_data:
            data.append(row_data)
        print('getting row infos ... ')

except Exception as e :
    print(e)
finally:
    print('scrapring table is finished\n data: \n',len(data))
    for row in data :
        print(row)
    if driver:
        driver.quit()

  



try :
    
    table = driver.find_element(By.ID,TABLE_ID)
    print('getting the table ... ')
    rows = table.find_elements(By.TAG_NAME , 'tr')
    for row in rows :
            
        cells = row.find_elements(By.TAG_NAME , 'td')+row.find_elements(By.TAG_NAME, 'td')
    
        row_data = [cell.text.strip() for cell in cells if cell]
        if row_data:
            data.append(row_data)
        print('getting row infos ... ')

except Exception as e :
    print(e)
finally:
    print('scrapring table is finished\n data: \n',len(data))
    for row in data :
        print(row)
    if driver:
        driver.quit()
```

## 3. Interacting with Elements
- `send_keys()` types into input boxes
- `click()` clicks buttons or links
- `clear()` clears text fields

```python
search_box = driver.find_element(By.NAME, 'q')
search_box.send_keys('Selenium tutorial')  # Type text
search_box.clear()  # Clear text
search_box.send_keys('Python Selenium')
search_box.submit()  # Submit form
```

## 4. Navigation
- `driver.get(url)` open a page
- `driver.back()` go back
- `driver.forward()` go forward
- `driver.refresh()` refresh page

```python
driver.get('https://books.toscrape.com/')
driver.back()
driver.forward()
driver.refresh()
```

## 5. Scrolling
- `driver.execute_script('window.scrollTo(0, document.body.scrollHeight)')` scrolls page

```python
# Scroll to bottom
driver.execute_script('window.scrollTo(0, document.body.scrollHeight)')
```

## 6. Getting Element Info
- `element.text` gets visible text
- `element.get_attribute('attr')` gets attribute value
- `element.tag_name` returns the tag name

```python
element = driver.find_element(By.TAG_NAME, 'h1')
print(element.text)
print(element.get_attribute('class'))
print(element.tag_name)
```

## 7. Waiting for Elements
- `time.sleep(seconds)` for simple wait
- `WebDriverWait` + `expected_conditions` for dynamic wait

```python
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

time.sleep(2)  # Wait 2 seconds

wait = WebDriverWait(driver, 10)
element = wait.until(EC.presence_of_element_located((By.ID, 'searchInput')))
```

## 8. Closing Browser
- `driver.close()` closes current tab
- `driver.quit()` closes whole browser

```python
driver.close()  # Close current tab
driver.quit()   # Close entire browser
```
