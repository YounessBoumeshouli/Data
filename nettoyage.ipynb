```python
import os
import pandas as pd
path = 'matchs_data'
fichiers = os.listdir(path)
for fichier in fichiers :

    df = pd.read_csv(path+'/'+fichier)
    df = df[df['Date'] != "Date"]
    df['Time'] = df['Time'].map(lambda x:x.split(' ')[0])
    df.dropna()
    df.drop_duplicates()
    df.to_csv(path+'/'+fichier,index=False)

```

```python
path = 'teams_data'
fichiers = os.listdir(path)
for fichier in fichiers :

    df = pd.read_csv(path+'/'+fichier)
    df = df[df['Pos'] != "Playing Time"]
    df = df[df['Pos'] != "Pos"]
    df.dropna()
    df.drop_duplicates()
    df.to_csv(path+'/'+fichier,index=False)
```
