```python
import os
import pandas as pd
path = 'matchs_data'
fichiers = os.listdir(path)
for fichier in fichiers :

    df = pd.read_csv(path+'/'+fichier)
    df = df[df['Date'] != "Date"]
    df.dropna()
    df.drop_duplicates()
    df.to_csv(path+'/'+fichier)

```

```python
import os
import pandas as pd
path = 'teams_data'
fichiers = os.listdir(path)
for fichier in fichiers :

    df = pd.read_csv(path+'/'+fichier)
    df = df[df['Age'] != "Performance" ]
    df = df[df['Age'] != "Age" ]
    df.dropna()
    df.drop_duplicates()
    df.to_csv(path+'/'+fichier)
```
